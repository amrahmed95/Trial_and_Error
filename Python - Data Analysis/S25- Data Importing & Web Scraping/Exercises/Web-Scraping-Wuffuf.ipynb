### Importing Libraries & Methods

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen 
```

### Inputting the URL

```python
url = 'https://wuzzuf.net/search/jobs/?a=spbg&q=illustrator'
```

### Create a client-based Request to get the URL

```python
client = urlopen(url=url)
```

### Getting the HTML Code of the full page

```python
html_page = client.read()
html_page
```

### Closing the Request

```python
client.close()
```

### Creating an HTML Parser Using BeautifulSoup

```python
soup = bs(html_page, 'html.parser')
soup
```

### Creating a Container for the Needed Data

```python
container = soup.find_all('div', {'class':'css-1gatmva e1v1l3u10'})
container
```

### Bringing it all together

```python
file = open('wuzzuf-illustrator.csv', 'w')
header = 'Job_title, Company, Job_Type\n'
file.write(header)
```

```python
for con in container:
    Job_title = con.find_all('h2', {'class':'css-m604qf'})
    Job_title = Job_title[0].text.strip()

    Company = con.find_all('a', {'class':'css-17s97q8'})
    Company = Company[0].text

    Job_Type = con.find_all('div', {'class':'css-1lh32fc'})
    Job_Type = Job_Type[0].text

    print(Job_title)
    print(Company)
    print(Job_Type)
    print()


    #data = Job_title + ',' + Company + ',' + Job_Type + '\n'
    #file.write(data)
```

```python
for con in container:
    Job_title = con.find_all('h2', {'class':'css-m604qf'})
    Job_title = Job_title[0].text.strip()

    Company = con.find_all('a', {'class':'css-17s97q8'})
    Company = Company[0].text

    Job_Type = con.find_all('div', {'class':'css-1lh32fc'})
    Job_Type = Job_Type[0].text

    data = Job_title + ',' + Company + ',' + Job_Type + '\n'
    file.write(data)
    
```
