### Joining Using Concat()
##### Merging two or more Datasets without restrictions / without conditions

##### Better to use Concat() when the datasets with same count of columns and same column names
##### May Duplication happen
###### Append() is removed

```python
men04.head(), men04.shape
```

```python
men08.head(), men08.shape
```

```python
# Concatenate DataFrames together
pd.concat([men04, men08])
```

```python
pd.concat([men04, men08], ignore_index=True)
```

```python
pd.concat([men04_det, men08_det], ignore_index=False, names=['year'], keys=['2004', '2008'])
```

```python
men_concat = pd.concat([men04_det, men08_det], ignore_index=False, names=['year'], keys=['2004', '2008']).reset_index()
men_concat
```

```python
pd.concat([men04_det, men08_det], ignore_index=False, names=['year'], keys=['2004', '2008']).reset_index().drop(columns='level_1')
```

### Inner Join (Intersection between Datasets)

```python
men_inner = men04.merge(men08, how='inner', on='Athlete', suffixes=['_2004', '_2008'], indicator=True)
men_inner
```

### Left Join ( Left_Only + Both)

```python
men_left = men04.merge(men08, how='left', on='Athlete', suffixes=['_2004', '_2008'], indicator=True)
men_left
```

### Right Join ( Right_Only + Both)

```python
men_right = men04.merge(men08, how='right', on='Athlete', suffixes=['_2004', '_2008'], indicator=True)
men_right
```
