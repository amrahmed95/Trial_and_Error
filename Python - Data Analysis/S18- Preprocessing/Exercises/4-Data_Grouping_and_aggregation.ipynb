#### Using agg() function

```python
titanic_simple = titanic.loc[:, ['survived','pclass','sex' ,'age','fare']]
titanic_simple
```

```python
titanic_simple.groupby('sex').mean()
```

```python
# Multiple Aggregation
titanic_simple.groupby('sex').agg(['mean', 'min', 'max' ,'sum'])
```

```python
titanic_simple.groupby('sex').agg({'survived': ['min', 'max'],
                                   'age': ['sum', 'count']
                                   })
```

```python
titanic_simple.groupby('sex').agg(survived_total = ('survived', 'sum'))
```

#### Dealing with Missing Data Using groupby()

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic['age'].mean()
```

```python
titanic.groupby(['sex','pclass'])['age'].mean()
```

```python
titanic['mean_age'] = titanic.groupby(['sex','pclass'])['age'].transform('mean')
titanic
```

```python
titanic['age'].fillna(titanic['mean_age'])
```

##### Q. Find the total number of medals for each city, and indicate the type of medals

```python
olympics.groupby(['City', 'Medal']).count()
```

```python
olympics.groupby(['City', 'Medal'])['Medal'].count()
```

```python
list(olympics.groupby('City'))
```

```python
olympics.groupby('City')['Medal'].count()
```

```python
olympics.groupby('City').agg(Medal_count = ('Medal', 'count'))
```

```python
olympics.groupby('City').agg(Medal_count = ('Medal', 'value_counts')).sort_values(by='Medal_count', ascending=False)
```

##### Q. Find the total number of participants and indicating their gender in each year

```python
olympics.head()
```

```python
olympics.groupby(['Edition','Gender'])['Athlete'].count()
```

```python
olympics.groupby('Edition')['Gender'].value_counts()
```
