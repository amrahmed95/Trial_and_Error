## Regular Expression Exercises

#### Check if String contains Only Defined Characters Using Regex

```python
s = '12dskxfm12343gsd'
if re.search(r'[123]', s):
    print('Valid')
else:
    print('Invalid')
```

#### Count Uppercase, Lowercase, and numeric values Using Regex

```python
s = 'Hello, my name is Mustafa Othman, Iam from Egypt.  I am 35 years old'
uppercase = re.findall(r'[A-Z]', s)
lowercase = re.findall(r'[a-z]', s)
numeric = re.findall(r'[0-9]', s)
len(uppercase), len(lowercase), len(numeric)
```

#### Use Regex to extract maximun numeric value from a string

```python
S = '100dsNDMSLAN21ASDKASH637HADSJHDSAJK12'
Max_Numeric_Value = re.findall(r'\d+', S)
Max_Numeric_Value
# ['100', '21', '637', '12']

# get max value 
max(Max_Numeric_Value)
```

#### Remove all characters except letter and numbers

```python
s = '100dsND$MSLAN!21AS#KASH6%37HADS^JHDSAJ:K12'
characters = re.findall(r'[\W]+',s)
characters #['$', '!', '#', '%', '^', ':']

re.sub(r'[\W]+', '', s)
# '100dsNDMSLAN21ASDKASH637HADSJHDSAJK12'
```

#### Regex to put spaces between words starting with capital letters

```python
name = 'MustafaOthmanMustafaOthmanSayed'
Names = re.findall(r'[A-Z][a-z]*',name)
# ['Mustafa', 'Othman', 'Mustafa', 'Othman', 'Sayed']

# Convert list to a string
Full_Name = ' '.join(Names)
Full_Name

```
