```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Show the distribution of the total bill with a histogram
sns.distplot(tips['total_bill'], bins=20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips,hue='sex', palette='coolwarm')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='rainbow')
```
