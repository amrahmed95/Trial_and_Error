```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
temp_cairo = [22.5, 22.7, 25, 27.3, 26.1, 22.4, 20]
temp_asyut = [30.2, 29, 25, 33.3, 39.7, 35, 28]
```

```python
df_cairo = pd.DataFrame({'days':days, 'temp_cairo':temp_cairo})
df_cairo
```

```python
df_Assyut = pd.DataFrame({'days':days, 'temp_asyut':temp_asyut})
df_Assyut
```

Line Plot

```python
plt.plot(df_cairo['days'], df_cairo['temp_cairo'], label='cairo')
```

Scatter Plot

```python
plt.scatter(df_cairo['days'], df_cairo['temp_cairo'])
```

Bar Charts

```python
plt.bar(df_cairo['days'], df_cairo['temp_cairo'])
```

Histogram

```python
my_data = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(my_data, bins=12)
```

Customizing Plots

```python
plt.plot(df_Assyut['days'], df_Assyut['temp_asyut'], label = 'Asyut', ls = '--', marker = 's', linewidth = 3, alpha = 0.5)
plt.plot(df_cairo['days'], df_cairo['temp_cairo'], label = 'Cairo', color = 'g', marker = 'o', markersize = 10)
plt.title('Temperature in Cairo and Asyut', fontsize = 15)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.grid()
plt.ylim(15,60)
#plt.legend(['Cairo', 'Asyut'])
plt.legend(loc = 0)

plt.savefig('temp.png', dpi = 300)
```

Matplotlib Subplots (Object-Syntax)

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(df_cairo['days'], df_cairo['temp_cairo'])
ax[0].set_title('Cairo')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Temperature')
ax[0].set_ylim(15,60)

ax[1].plot(df_Assyut['days'], df_Assyut['temp_asyut'])
ax[1].set_title('Asyut')
ax[1].set_xlabel('Days')
ax[1].set_ylabel('Temperature')
ax[1].set_ylim(15,60)
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1 ,0.1 , 0.8 ,0.8])
ax.plot(x,y, linestyle='--', color='r', marker='o')
ax.set_title('Power of 2')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1 ,0.1 , 0.8 ,0.8])
ax2 = fig.add_axes([0.2 ,0.5 , 0.4 ,0.3])

ax1.plot(x,y)
ax1.set_title('Power of 2')

ax2.scatter(y,x)
ax2.set_title('Power of 2')
```
