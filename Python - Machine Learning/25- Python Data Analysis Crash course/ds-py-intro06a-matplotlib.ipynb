# Matplotlib Library

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

# Plot Data

```python
days = [1,2,3,4,5,6,7]
temp_cairo = [22.5, 22.7, 25, 27.3, 26.1, 22.4, 20]
temp_asyut = [30.2, 29, 25, 33.3, 39.7, 35, 28]

df_cairo = pd.DataFrame(data = {'Day': days, 'Temperatures': temp_cairo})
df_asyut = pd.DataFrame(data = {'Day': days, 'Temperatures': temp_asyut})
```

```python
df_cairo
```

```python
df_asyut
```

## Line Plots

```python
plt.plot(df_cairo['Day'], df_cairo['Temperatures'])
plt.show()
```

```python
plt.plot(df_asyut['Day'], df_asyut['Temperatures'])
plt.show()
```

## Scatter Plots

```python
plt.scatter(df_cairo['Day'], df_cairo['Temperatures'])
plt.show()
```

## Bar Charts

```python
plt.bar(df_asyut['Day'], df_asyut['Temperatures'])
plt.show()
```

## Histograms

```python
my_data = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(my_data, bins = 5)
plt.show()
```

## Customizing Plots

```python
plt.plot(df_cairo['Day'], df_cairo['Temperatures'], label = 'Cairo', marker = 's', ls = '--')
plt.plot(df_asyut['Day'], df_asyut['Temperatures'], label = 'Asyut', c = 'r', alpha = 0.4)

# x-axis & y-axis
plt.xlabel('Days of the Week')
plt.ylabel('Temperatures in Cel.')

# title
plt.title('Temperatures Degrees in Cairo')

# plot titles
plt.legend()

plt.show()
```

## Subplots

```python
figure, axes = plt.subplots(1, 2, figsize = (10,5))

axes[0].plot(df_asyut['Day'], df_asyut['Temperatures'], c = 'r')
axes[0].set_title('Temperatures Degrees in Asyut')

axes[1].plot(df_cairo['Day'], df_cairo['Temperatures'], c = 'k')
axes[1].set_title('Temperatures Degrees in Cairo')

plt.savefig('temp.png')

plt.show()
```
