#### Data Preprocessing

```python
# Transaction is a list of list => Nested for loop

transactions = []

for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
transactions[:1]
```

#### Training Apriori on Dataset

```python
!pip install apyori     # apyori is build by python Foundation
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)

# min_length => at least two products with each other (A , B)

# These parameters are very experimental
#-------------------------------------------
# To determine min_support:
# Let's say some products are bought 3 times daily so =>  (3 times * 7 days) / 7501 = 0.0028
#-------------------------------------------
# Tp determine min_confidence , min_lift
# Search in Forums like Kaggle for a similar situation (min_support = 0.003 & similar dataset )  
```

#### Visualizing the results

```python
results = list(rules)
```

```python

results_list = []
for i in range(0, len(results)):
    results_list.append('RULE: ' + str(results[i][0]) + ' ==> SUPPORT: ' + str(results[i][1]) + ' ==> CONFIDENCE: ' + str(results[i][2][0][2]) + ' ==> LIFT: ' + str(results[i][2][0][3]))

results_list
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
